### 2.- Generar una onda cuadrada aproximando hasta el k- 칠simo t칠rmino, utilizando exclusivamente operaciones con matrices y funciones matriciales de numpy, en otras palabras, no est치 permitido utilizar ning칰n tipo de ciclos. Graficar el resultado. Asuma una frecuencia de muestreo de 300 Hz y una frecuencia fundamental de 5 Hz.

```python
# Importamos las librerias
import numpy             as np
import matplotlib.pyplot as plt
```

```python
# s = A/2 + 2A/pi*sen(t*w) + 2A/(3*pi)*sin(3*t*w) + ...
# Donde: 
# n  = numero de coef. para aproximar
# A  = amplitud
# L  = tama침o de t
# f  = frecuencia fundamental
# sr = frecuencia de muestreo

def aprox_cuad (n = 21, A=2,  L = 25, f = 5, sr = 300):
    t = np.linspace(0,L,L*sr)
    w = 2*np.pi/f
    temp = []
    temp.append(np.linspace(1,n,n//2+1))
    temp = np.array(temp).T
    x = temp*t
    x = np.sin(x*w)

    x1 =2*A/(np.pi*temp)
    
    x = x*x1
    x = np.sum(x,0)    
    s = A/2 + x
    plt.figure(figsize=(15, 4))
    plt.plot(t,s)
```

### 3.- Repita el ejercicio anterior, pero considerando una onda triangular.

```python
import numpy             as np
import matplotlib.pyplot as plt
```

```python
# s = A/2 - 4*A/pi^2*sen(t*w) + 4A/(3^2*pi^2)*sin(3*t*w) + ...
# Donde: 
# n  = numero de coef. para aproximar
# A  = amplitud
# L  = tama침o de t
# f  = frecuencia fundamental
# sr = frecuencia de muestreo

def aprox_tri (n = 21, A=2,  L = 25, f = 5, sr = 300):
    t = np.linspace(0,L,L*sr)
    w = 2*np.pi/f
    temp = []
    temp.append(np.linspace(1,n,n//2+1))
    temp = np.array(temp).T
    x = temp*t
    x = np.cos(-x*w)

    x1 =4*A/((np.pi**2)*(temp**2))
    
    x = x*x1
    x = np.sum(x,0)    
    s = A/2 + x
    plt.figure(figsize=(15, 4))
    plt.plot(t,s)
```
